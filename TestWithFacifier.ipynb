```python
import cv2
import matplotlib.pyplot as plt
```

```python
emotions = ["afraid", "angry", "disgusted","happy",  "neutral", "sad","surprised","Contempt"]
```

```python
def detect_faces(img):
    facecascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
    faces = facecascade.detectMultiScale(img,scaleFactor=1.1, minNeighbors=5)
    return faces

# resize the image to 350x350 pixels
def resize_face(face):  
    face = cv2.cvtColor(face, cv2.COLOR_BGR2GRAY)
    face = cv2.resize(face, (350, 350))
    return face
```

```python
image = cv2.imread('SurpriseFace.jpg')
#image = cv2.imread('face-showing-happiness-microexpression-800x950.jpg')
```

```python
# detect faces
faces = detect_faces(image)

# for recognizing the emotion on the face
emotion_recognizer = cv2.face.FisherFaceRecognizer_create()
emotion_recognizer.read('emotion_classifier_model.xml')

# for recognizing the gender of the face
gender_recognizer = cv2.face.FisherFaceRecognizer_create()
gender_recognizer.read('gender_classifier_model.xml')
```

```python
for face in faces:
    # find the bounding box of the face
    (x, y, w, h) = face

    # draw rectangles around the face
    cv2.rectangle(image, (x, y), (x+w, y+h), (255, 0, 255), 2)

    # get the face
    roi_face = image[y:y+h, x:x+w]

    # predict emotion
    predicted_emotion = emotion_recognizer.predict(resize_face(roi_face))  
    emotion = emotions[predicted_emotion[0]]
    #print(f'emotion =  {predicted_emotion}')
    print(f'emotionType =  {emotion}')
    text = emotion

    # predict gender
    predicted_gender = gender_recognizer.predict(resize_face(roi_face))
    #print(f'gender  {predicted_gender}')   
    print(f'gender = {"female" if predicted_gender[0] == 0 else "male"}')        
    gender = "female" if predicted_gender[0] == 0 else "male"
    text += ", " + gender

    # display the emotion and gender
    cv2.putText(image, text, (x,y-10), 
                cv2.FONT_HERSHEY_SIMPLEX, 1, (0,255,255), 2)
       
# show the frame
cv2.imshow("Image", image)

# cleanup
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.waitKey(1)
```
